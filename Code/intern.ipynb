### 2. Data Preprocessing

create a new column called 'set_1_2_3' to see if each person has 1,2,3 as the set (only focus on 1-5)

```python
set1 = {1,2,3}

def f(row):
    string2 = row['set_126'].replace('[', '').replace(']', '').replace(' ', '')
    set2 = set([int(s) for s in string2.split(',')]) # convert string to list then to set
    set3 = {x for x in set2 if x <= 5} # only keep the number 1-5
    if set1 == set3:
        return 1
    else:
        return 0
    
df_test['set_1_2_3'] = df_test.apply(f, axis = 1)

df_test[df_test['set_1_2_3'] == 1]
```

```python
df_control['set_1_2_3'] = df_control.apply(f, axis = 1)

df_control[df_control['set_1_2_3'] == 1]
```
